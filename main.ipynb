```python
import matplotlib.pyplot as plt
import numpy as np

xpoints1 = np.array([2, 2.5, 3.7, 4.5, 4.8, 5, 6])
ypoints1 = np.array([4, 5, 2, 9, 3, 10, 12])
plt.subplot(2, 1, 1)
plt.plot(xpoints1, ypoints1, ls = ':', c = 'g', marker = 'o', ms = 10, mec = 'r', mfc = 'k')

xpoints2 = np.array([3, 5, 7, 9, 2, 4])
ypoints2 = np.array([8, 10, 2, 6, 1, 5])
plt.subplot(2, 1, 2)
plt.plot(xpoints2, ypoints2, ls = ':', c = 'b', marker = 'd', ms = 12, mec = 'k', mfc = 'r')

plt.title("stress-strain graph", loc = 'left')
plt.xlabel("stress",)
plt.ylabel("strain")
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

target = 50
lhl = 40
rhl = 60

def fuzzzyfier(speed):
     if(speed <= lhl):
       return 0
     if(speed >= rhl):
      return 1
     if(speed > lhl & speed < target):
       return (speed - lhl)/(target - lhl)
     if(speed < rhl & speed > target):
      return (speed - rhl) / (target - rhl)
         
speeds = np.arange(10, 100, 0.5)
y = []

for speed in speeds:
       y.append(fuzzzyfier(speed))

y = np.asarray(y)
fig, (ax1) = plt.plot(1, 2, figsize = (10, 4))
ax1.plot(speeds, y, color = "red")
ax1.set(xlabel = "speed", ylabel = "membership")
plt.show()
     
```
